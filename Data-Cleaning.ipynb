```python
import pandas as pd
import numpy as np
```

```python
nfl_data = pd.read_csv("NFL Play by Play 2009-2017 (v4).csv")
np.random.seed(0)
```

```python
nfl_data.head()
```

```python
nfl_missing_value = nfl_data.isnull().sum()
nfl_missing_value
```

```python
total_cell = np.product(nfl_data.shape)
total_missing = nfl_missing_value.sum()

(total_missing/total_cell) * 100
```

```python
nfl_missing_value[0:10]
```

# Drop NFL missing data

```python
nfl_data.dropna()
```

```python
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
print(f"column in original dataset: {nfl_data.shape[1]} \n")
print(f"column in dropped dataset: {columns_with_na_dropped.shape[1]} \n")
```

# Fill The Missing Value

```python
subset_nfl_data = nfl_data.loc[:, 'EPA' : 'Season']
subset_nfl_data
```

```python
subset_nfl_data.fillna(0)
```

```python
subset_nfl_data.fillna(method = 'bfill', axis = 0).fillna(0)
```
